```python
start = 'W'
endereco = {}
for place in range(0, 16):
    (endereco['{0:04b}'.format(place)]) = '00000000'
while start == 'W' or start == 'L' or start == 'R':
    start = input('Digite W para escrever, R para ler, L para listar toda a memoria e qualquer tecla para parar. ')
    if start == 'W':
        local = input('Digite o endereco de 4 bits: ')
        palavra = input('Digite o dado de 8 bits: ')
        endereco[local] = palavra
    elif start == 'R':
        local = input('Digite o endereco: ')
        print(endereco.get(local))
    elif start == 'L':
        print('Todos os estados: ')
        for element in endereco:
            print(endereco[element])
```
